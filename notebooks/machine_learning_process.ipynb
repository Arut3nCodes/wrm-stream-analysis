```python
import pandas as pd

wrm_df = pd.read_csv("../data/processed_wrm_data.csv")
owa_df = pd.read_csv("../data/owa_processed_data.csv")

wrm_df['timestamp'] = pd.to_datetime(wrm_df['timestamp'])
owa_df['timestamp'] = pd.to_datetime(owa_df['timestamp'])

merged_df = pd.merge(wrm_df, owa_df, on="timestamp", how="left")
```

```python
merged_df.isna().sum()

nan_rows = merged_df[merged_df[[
    "temperature", "feels_like", "pressure", "humidity",
    "dew_point", "clouds", "wind_speed", "wind_deg",
    "weather_main", "weather_description", "rain", "snow"
]].isna().any(axis=1)]

nan_rows[["timestamp"]].drop_duplicates().sort_values("timestamp")
```
