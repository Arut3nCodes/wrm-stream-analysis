### WRM Data collection description

Dane pobierane są z get ```https://api.nextbike.net/maps/nextbike-live.json?city=(city_id)```
gdzie city_id dla Wrocławia to 148

API nextbike w odpowiedzi zwraca typ pliku json z danymi miast, wszystkich stacji oraz rowerów dostępnych w systemie.
Dodanie parametru city zwraca powoduje zwrot danych tylko dla miasta o podanym id.

Dane przechowywane są w dwóch miejscach. W folderze wrm, w osobnych plikach dla każdego wykonanego requesta, przechowywane są nieprzetwarzone pobrane dane w formacie json. W pliku processed_wrm_data przechowywane są dane przetworzone 

W mojej analizie interesują mnie poniższe parametry:
- places - lista obiektów stacji z rowerami
- places.uid - id stacji
- places.lat + places.lng - współrzędne geograficzne stacji
- places.bikes - lista rowerów dostępna w stacji (parametry poszczególnych rowerów są dla mnie nieważne)
- places.available_bikes - dostępne rowery 

Poniżej użycie data collectora dla wrm:

````python
from data_collector import collect_data

bike_data = collect_data(bike_data_api, {"city": 148})
````

### Dane pogode - OpenWeatherMap

Teraz przejdziemy do analizy procesu pobierania danych pogodowych

Dane pogodowe pobierane są za pomocą One Call API 3.0 udostępnianego przez firmę OpenWeatherMap

Udostępniane są dane pogodowe z
    obecnymi odczytami ze stacji pogodowej
    minutowymi predykcjami
    godzinnymi predykcjami  
    dziennymi predykcjami

Na ten moment interesują mnie tylko dane z obecnymi odczytami, jednak w celach generowania predykcji być może zainteresuję się w przyszłości cogodzinnymi predykcjami pogody

Struktura żądania:

```https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={lon}&units={metric}&appid={API key}```

gdzie:
* lat - szerokość geograficzna
* lon - długość geograficzna
* units - system jednostek w którym podawane są dane
* appid - klucz api udostępniany użytkownikowi w celu identyfikacji żądań

Współrzędne geograficzne dla Wrocławia to (51.1, 17.0333)
System jednostek oczywiście wybrałem metryczny
Klucz api przechowywany jest w zmiennej środowiskowej w systemie (na serwerze musiał zostać zhardcodowany bo nie wykrywał zmiennej środowiskowej)

Dane zawarte w żądaniu
